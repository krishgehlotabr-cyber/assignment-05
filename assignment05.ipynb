```python
products = pd.DataFrame({
    'Product':['Laptop','Mobile','Shirt','Headphones','Watch'],
    'Category':['Electronics','Electronics','Clothing','Electronics','Accessories'],
    'Price':[50000,15000,1200,2500,3000],
    'Quantity':[5,10,20,8,12]
})

print(products)
```

```python
print(products[products['Category']=='Electronics'])
```

```python
print(products[products['Price']>10000])
```

```python
print(products['Price'].mean())
```

```python
print(products.loc[products['Price'].idxmax()])
```

```python
print(products.loc[products['Price'].idxmin()])
```

```python
print(products[products['Quantity']>7])
```
